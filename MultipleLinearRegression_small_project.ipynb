```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
x=pd.DataFrame(fetch_california_housing().data,columns=fetch_california_housing().feature_names)
```

```python
x
```

```python
y=pd.DataFrame(fetch_california_housing().target,columns=['target'])
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
sc_x=StandardScaler()
x_train_scaled=sc_x.fit_transform(x_train)
x_test_scaled=sc_x.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression(fit_intercept=True,positive=False,n_jobs=None,copy_X=True)
```

```python
model.fit(x_train_scaled,y_train)
```

```python
y_pred=model.predict(x_test_scaled)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
r2_score(y_test,y_pred)
```

```python
mse=mean_squared_error(y_test,y_pred)
```

```python
mse
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
rmse=np.sqrt(mse)
rmse
```
